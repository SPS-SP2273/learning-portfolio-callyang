Task 4

```python
toberenamed = glob.glob('os-collaboration-exercise-data/*.txt', recursive = True)
# to get all the files in the os-collaboration-exercise-data
# ** indicates all folders inside, but no need cause no folders are inside this folder
# .txt to get all text files otherwise i will rename everything

def rename_my_file(old_file_name):
    for file in old_file_name:

        filename = os.path.split(file)[1] # gets only the file name
        month = filename[0:3] # gets just the 3 letter month
        date = filename[4:6]
        location = filename[7:]
        oldpath = file
        newname = f'{monthnumbers.get(month)}-{date}-{location}'
        newpath = os.path.join('os-collaboration-exercise-data', newname) 
        os.rename(oldpath, newpath)

rename_my_file(toberenamed)
```

Task 7

```python
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

def sortcity(chosencity):
    citypath = os.path.join('os-collaboration-exercise-data', chosencity)
    lowercasecity = chosencity.lower()
    citydata = glob.glob(f'os-collaboration-exercise-data/*{lowercasecity}.txt*')
    for cityfiles in citydata:
        shutil.copy(cityfiles, citypath)

for city in cities:
    sortcity(city)
```
