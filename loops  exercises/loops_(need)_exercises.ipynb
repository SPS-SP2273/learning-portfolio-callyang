```python
temperature_celsius = [
    0,5,10,15,20,25,
    30,35,40,45,50
]

farenheitlist = []
for temperature in temperature_celsius:
    farenheit = temperature*1.8 + 32
    farenheitlist.append(farenheit)
    
print(f'list of temperatures in Farenheit is {farenheitlist}')
print(f'list of temperatures in Celsius is {temperature_celsius}')
```

```python
# for ___ in ___ can be string also
for letter in ['A','B','C']:
    print(letter)
    for number in[1,2,3]:
        print(f'{letter}{number}', end= '\t')
    print('\n')
```
