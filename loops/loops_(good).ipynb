# 2 List comprehension!

## 2.1 Basic syntax

```python
# creating a fresh list from a list
[number for number in range(5)] # a list called number 
```

```python
[number**2 for number in range(5)]
```

## 2.2 List comprehension with conditions

```python
[number for number in range(10) if number %2 ==0]
# u can even use condition
```

```python
[f'hehe {power}' for power in range(10) if power%2==0]
```

# 3 Other useful stuff

## 3.1 for with unpacking

```python
# unpacking
x, y, z = [1,2,3]
print(x,y,z)
print(f'x = {x}, y = {y}, z = {z}')
```

```python
py_superhero_info = [['Natasha Romanoff', 'Black Widow'],
                     ['Tony Stark', 'Iron Man'],
                     ['Stephen Strange', 'Doctor Strange']]

for x, y in py_superhero_info: # for doesnt need to be only
                               # one thing, for x in __ eg
                               # can be x,y in list
    print(f"{x} is Marvel's {y}!")
```

## 3.2 for with zip()

```python
super_names = ["Black Widow", "Iron Man", "Doctor Strange"]
real_names = ["Natasha Romanoff", "Tony Stark", "Stephen Strange"]

for x, y in zip(real_names, super_names): # combine 2 list
    print(f"{x} is Marvel's {y}!")        # zip doesnt do anyth until u use it
```

```python
list(zip(real_names, super_names))
```

## 3.3 for with dictionaries

```python
cakerating = {"Strabwerry": "Super good",  # pls rmb to separate w comma
             "Chocolate": "Very good",
             "Cheese": "Bad"}

for key,value in cakerating.items():
    if key == "Cheese":
        print(f"{key}cake is {value}")
    else:
        print(f"{key} cake is {value}")
```
