```python
def convert_farenheit(temperature_farenheit,target_scale = 'Celsius'):
    temperature_farenheit = str(temperature_farenheit)
    temperaturecheck = temperature_farenheit.replace('.','') # so that isnumeric can check
    
    if temperaturecheck.isnumeric() == False:
        print("Input is not a number")
        answer = "Error! Non number given/decimal number given."
        print(f'The converted temperature is {answer}, units {target_scale}')
        
    else:
        print("Input is a number")
        if target_scale == 'Celsius':
            answer = (float(temperature_farenheit)-32)/1.8
            print(f'The converted temperature is {answer}, units {target_scale}')
        elif target_scale == 'Kelvin':
            answer = (float(temperature_farenheit)-32)/1.8+273.15
            print(f'The converted temperature is {answer}, units {target_scale}')
        else:
            answer = "Error in units!"
            target_scale = 'invalid'
            print(f'The converted temperature is {answer}, units {target_scale}')
```
