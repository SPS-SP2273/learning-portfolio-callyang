## 1.2 Anonymous functions

```python
def add(j):
    return j+5
```

```python
add_v2 = lambda x: print(x+5) # lambda only one line
                              # automatically puts return,, and stores value
                              # if u manually put print, if u print out with the newly defined function
                              # it will be none
```

```python
print(add_v2(5))
```

```python
shortfunc = lambda name: f"Hey {name}, how are you?"
                  # ^ the argument   ^ the output
```

```python
shortfunc(name = 'Cally')
# lambda function always returns value of last statement
```

```python
test_list = ['q', 'c', 'g']
sorted(test_list)
```

```python
numbers=[[9, 0, -10],
         [8, 1, -11],
         [7, 2, -12],
         [6, 3, -13],
         [5, 4, -14],
         [4, 5, -15],
         [3, 6, -16],
         [2, 7, -17],
         [1, 8, -18],
         [0, 9, -19]]
```

```python
sorted(numbers) # it takes each index 0 of each list inside, and ranks it 
```

```python
sorted(numbers, key=lambda x: x[1]*x[0]-x[2])
               # for x in numbers:
               #     x[1]*x[0]-x[2]
               # but it does not change the value in it
```

## 1.3 Optional arguments

```python
def greeting(name =', you forgot to put a name'):
    if name == 'Batman':
        return 'Omg its Batman!:O'
    else:
        return f'Hey {name}!'
```

```python
greeting() # will take the '' as default name and print 
           # under the else statement!
```
