## 2.1 Positional, keyword and default arguments

```python
def sidebyside(a,b,c=42):
    return f'{a}|{b}|{c}' # i dont care what number it is just print out
```

```python
def sidebyside(a,b,c=42):
    return f'{a:2d}|{b:d}|{c:5d}' # padding 
```

```python
print(sidebyside(1,2,3)) # now 3 is forced to take 5 spaces
print(sidebyside(1,2,1234))
```

```python
sidebyside(c=2,b=4,a=6) # order does not matter cause u specify
```

```python
sidebyside(3,4) # did not define c but they will use the default
                # keywords mean when u use a =, sep = etc
                # positional means relying on defined position
```

```python
sidebyside(a=2,1) # will not work because python dont know where is 1
```

## 2.2 Docstrings

```python
# explanation of what the function does
# will show if we do help()

def sidebyside(a,b,c = 42):
    '''
    A test function to demonstrate how positional, keyword and default arguments work.
    '''
    return a*2,c*2,b*2  # must have comma
                        # return statement give back 1 thing
                        # to give more than one, need comma
                        # output will follow return statement

sidebyside(2,3,4)
```
