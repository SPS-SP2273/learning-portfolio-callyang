# 2 Comparing the two ‘dialects’

```python
from matplotlib import pyplot as plt

xarray = np.linspace(-np.pi,np.pi,num =100)
cosx = np.cos(xarray)
sinx = np.sin(xarray)

plt.plot(xarray, cosx, label = 'cos x')
plt.plot(xarray, sinx, label = 'sin x')
plt.legend()

##### OO interface
fig, ax = plt.subplots(nrows = 1, ncols = 1) 
# use plt.subplots to create a figure and axis
# how many rows and columns uw to split the cell output

ax.plot(xarray, cosx, label = 'cos x')
ax.plot(xarray, sinx, label = 'sin x')
ax.legend()
```

# 4 A complete OO example

```python
fix, ax = plt.subplots(nrows = 2, ncols = 1, 
                       figsize = (5,8),sharex = True) # specify dimension of figures, 
                                                      # sharex to share x axis
ax[0].plot(xarray, cosx, label = r'$\cos(x)$')
ax[0].fill_between(xarray, 0, cosx, alpha = .25)
ax[1].plot(xarray, sinx, label = r'$\sin(x)$') # shades figure 'under' graph
ax[1].fill_between(xarray, 0, sinx, alpha = .25)

for a in ax: # so these are applied to both ax[0] and ax[1]
    a.legend()
    a.grid(alpha = .25)
    a.set_ylabel('$y$', rotation = 0)
    
ax[1].set_xlabel('$x$') # since we are sharing the x axis, just label the lowest plot

fig.suptitle(r'$\sin(x)$ and $\cos(x)$') # title of figure not plot
fig.tight_layout() # look nice
```

# 5 Multiple rows and columns

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2,
                      figsize = (5,5), sharex = 'col', sharey = True)

top_left, top_right, bottom_left, bottom_right = ax.flatten()  
# because now ax is a 2 by 2 array
# flatten takes the 2D array and flattens it into 1D

top_left.set_title('top Left')
top_right.set_title('top Right')
bottom_left.set_title('bottom Left')
bottom_right.set_title('bottom Right')

# setting the horizontal lines
top_left.hlines(y = 1, xmin = 0, xmax = 4)
top_right.hlines(y = 2, xmin = 0, xmax = 5)
bottom_left.hlines(y = 3, xmin = 0, xmax = 4)
bottom_right.hlines(y = 4, xmin = 0, xmax = 5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

### Using ax

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2,
                      figsize = (5,5), 
                      sharex = 'col', sharey = 'row')

ax.shape # 2 by 2 array
# hence can access each of the plots 
ax[0,0].set_title('top left')
ax[0,1].set_title('top right')
ax[1,0].set_title('bottom left')
ax[1,1].set_title('bottom right')

# also can
# top_left = ax[0,0]
# top_right=ax[0, 1]
# bottom_left=ax[1, 0]
# bottom_right=ax[1, 1]

# by doing flatten, u are creating a 1D array, can directly equate to the names
top_left, top_right, bottom_left, bottom_right = ax.flatten()
```

### A quick helper function

```python
def get_plot_data():
    data = {} # creating an empty dictionary
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_all.csv'
    data['ALL DATA'] = np.loadtxt(filename, skiprows =1)
    
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_nus.csv'
    data['NUS DATA'] = np.loadtxt(filename, skiprows =1)
    
    return data

get_plot_data()
```

### The histogram

```python
plt.style.use('bmh')
data = get_plot_data()

# bins specify how many bins to split data into
plt.hist([data['ALL DATA'], data['NUS DATA']], bins=50, label = ['All','NUS'])
plt.xlabel("Mean of Basic Monthly Earning (S$)")
plt.ylabel("Number of Students")
plt.legend()
```

## 6.2 Scatter plots

```python
plt.style.use("seaborn-v0_8-darkgrid")

data = get_plot_data()

for label, numbers in data.items(): # .items() give key value pair
    x = range(len(numbers))
    y = numbers
    plt.scatter(x, y, label=label, alpha=.5)

plt.xlabel('Position in the list')
plt.ylabel('Mean of Basic Monthly Earning (S$)')
plt.legend()
```

## 6.3 Bar charts

```python
# barh() if horizontal bars
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys()) 
numbers = list(student_numbers.values())

plt.style.use('ggplot')
plt.bar(majors, numbers) # specify x, y values
plt.xlabel('Majors')
plt.ylabel('Number of Students')
```

## 6.4 Pie charts

```python
cake_sold = {'Strawberry': 14,'Cheese': 9,
            'Chocolate': 10,'Blueberry': 3}

flavour = list(cake_sold.keys())
sold = list(cake_sold.values())

plt.style.use('fivethirtyeight')
plt.pie(sold, labels = flavour,
        autopct='%1.1f%%',   # How to format the percentages
        startangle=-90                
        )
plt.title('Percentage of each flavour sold')
```
