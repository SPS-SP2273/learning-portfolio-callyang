# 1 A simple plot

## 1.1 Letâ€™s look at some code.

```python
x = [0,1,2,3,4,5,6]
y_1 = [0,2,4,6,8,10,12]
y_2 = [0,3,6,9,12,15,18]

plt.plot(x, y_1, label = 'Y values', color = 'red', linestyle = 'dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('y against x')
plt.grid(alpha = .1) # how faint the grid lines are
plt.legend(loc = 'upper left')
```

### Things to note

```python
# Abbreviations
# color c
# linestyle ls
# linewidth lw

# plt.plot(x, y, color='red', linestyle='dashed', linewidth=2) is the same as
# plt.plot(x, y, c = 'red', ls = 'dashed', lw=2)
```

```python
# jupyter notebook is interactive hence will give plot without 
# plt.show()
# but for code to work in non-interactive env need plt.show()

plt.plot(x, y_1,y_2) # side by side
```

## 1.3 Yet another plot but with error bars

```python
# using plt.errorbar() 

y_5 = [2,4,5,6,7,4,2]
x_error, y_error = 0.1, 0.5

plt.plot(x, y_3, label = 'Y3 values', color = 'red', ls = 'dashed')
plt.plot(x, y_4, label = 'Y4 values', color = 'blue')
plt.errorbar(x,y_5, xerr = x_error, yerr = y_error, label = 'Y3 with error', color = 'green')

plt.xlabel('x values')
plt.ylabel('Y3/Y4 values')
plt.title('X vs Y')
plt.grid(alpha = .25)
plt.legend(loc = 'lower right')
```

## 2.1 Adding mathematical functions

```python
# numpy array allow for data related math functions
# like x square, sin x

x2 = xarray**2
sinx = np.sin(xarray) # looks bad because sinx values is -1 to 1

plt.plot(xarray, x2, label = 'x^2', color = 'green', ls = 'dashed')
plt.plot(xarray, sinx, label = 'sin(x)', color = 'purple')
plt.legend() # default upper left
```

## We need another axis!

```python
# twinx() for 2 y axis
xarray = np.linspace(0,6,100) 
x2 = xarray**2
sinx = np.sin(xarray)

plt.plot(xarray, x2, label = 'x^2', color = 'red', ls = 'dashed')
plt.legend(loc = 'lower left')

plt.twinx()

plt.plot(xarray, sinx, label = 'sin(x)', color = 'blue')
plt.legend(loc = 'lower right')
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
temp = np.loadtxt('Complete_TAVG_daily.txt', skiprows = 24) # skip first 24 lines

date = temp[:,0]
anomaly = temp[:,-1] # gives Anomaly column
print(date)
print(anomaly)

plt.plot(date, anomaly, alpha = 0.5)
plt.ylim([-8,8])

plt.hlines(0, date[0], date[-1], ls = '--', colors = 'grey')
#         ^ y value ^ start and end values of x, basically span all the dates
```

## 4.2 Styles

```python
plt.style.use('fivethirtyeight')

temp = np.loadtxt('Complete_TAVG_daily.txt', skiprows = 24) # skip first 24 lines

date = temp[:,0]
anomaly = temp[:,-1]
print(date)
print(anomaly)

plt.plot(date, anomaly, alpha = 0.5)
plt.ylim([-8,8])

plt.hlines(0, date[0], date[-1], ls = '--', colors = 'grey')

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temporature anomaly\n(Relative to average from Jan 1951 - Dec 1980)')
```
