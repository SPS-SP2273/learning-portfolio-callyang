# 6 Combining English and variables

```python
name = 'Batman'
print(f"Hello {name}!")

# print('Hello' + name) would also work
```

```python
x = 20
print(f"The value of {x} squared is {x**2}")
```

```python
text = 'Strawberry ice cream is nice!'
print(text)

# just the regular print would work
# but f string allows for formatting

print(f'{text}')
```

```python
print(f"{text:>50}") #align right
                     # the number means a block of 50 char
```

```python
print(f"{text:<50}") # must use {} 
                     # aligh left
```

```python
print(f'{text:^30}')
```

```python
print(f'The cube of pi to 6 decimal places is {np.pi**3:.6f}')
      
# remember to end the whole thing with ' at the end
# between the number (pi cube) and declaring the decimal places
# put a :
```

```python
print(f'The cube of pi to 6 decimal places is {np.pi**3:6e}')
```

```python
print('The cube of pi to 6 decimal places is np.pi**3.6f')

# will not work if u dont use f string, can't declare decimal 
```
