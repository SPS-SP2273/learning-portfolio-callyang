## Quadratic Equations

### Introduction


![](https://en.wikipedia.org/wiki/File:Polynomialdeg2.svg)

*(Image from the [Wikipedia page on Quadratic equations](https://en.wikipedia.org/wiki/File:Polynomialdeg2.svg))*

The general form of a quadratic equation is:

$$
ax^2 + bx + c = 0
$$




### Solutions


Provided $ a \ne 0$, we can use an elementary algebraic method called [completeing the square](https://en.wikipedia.org/wiki/Completing_the_square) to show that a quadratic equation has the following solution:


$$
x = \dfrac{-b \pm \sqrt{b^2-4ac}}{2a}
$$

### Discriminant

The quantity $ \Delta = b^2 - 4ac$ is called the **discriminant** of the equation and decides the nature of its solutions. The table below shows the various possibilities.

| Discriminat|Roots|
| :---|---:|
| $ \Delta = b^2 - 4ac = 0$ | A single solution of $-b/(2a)$ |
| $ \Delta = b^2 - 4ac \gt 0$ | Two distinct solutions |
| $ \Delta = b^2 - 4ac \lt 0$ | No real solutions; both are complex |

5. A pull is when changes that are made on the GitHub webpage and committed is applied to the version of the files locally on the machine.
