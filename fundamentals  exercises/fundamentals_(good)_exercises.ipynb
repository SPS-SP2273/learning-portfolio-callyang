```python
score_input = input("Enter the student's score:")

if score_input.isnumeric() == False :  # although input is made into string
                                       # .isnumeric can still check
    grade = 'Invalid as score is not a number/ an integer'
    
else:
    score = int(score_input)     # input() makes the input a string
                                 # even when it is a number
    if 0 <= score <= 34:
        grade = 'Fail'           # this block of code should
                                 # be placed under else, not on its own
    elif 35 <= score <= 49:      # because otherwise it'd run 
        grade = 'C'              # regardless its a number or not

    elif 50 <= score <= 69:
        grade = 'B'

    elif 70 <= score <= 100:
        grade = 'A'

    else: 
        grade = 'Invalid as it is not within range of scores from 0-100'

print(f"The student's letter grade is: {grade}")



# .isnumeric checks if a variable is a string or a number
# should check score_input, not score
# int() can convert only if its eg. 23 (but made to string due to input())
# but if score_input given was a number, the conversion to int will fail
# and hence the checking of score.isnumeric will never run

```
