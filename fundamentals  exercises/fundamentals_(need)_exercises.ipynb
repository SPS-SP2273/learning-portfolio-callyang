|# | Term | Description
|:---|:--:|---:|
|1 | Function | An action you wish for python to run, for example print() or split()
|2 | Argument | A value passed to the function, for example 'hello'
|3 | Comments | A note to remember the purpose of the code
|4 | Indentations | Spaces that are crucial in python, such as in if statements
|5 | Packages | Files downloaded into the local system that includes specific code and extensions
|6 | . | Indicates ownership, that this function is owned by this package
|7 | : | Used to create dictionary list by assigning string to a name, or in if statements
|8 | [] | Used to create lists or to call out an element in a list, if used after a named list
|9 | [[[]]] | Used to indicate a list within a list
|10 | {} | Used to indicate dictionary lists

```python
test_list = [[[1,2,3]]]

test_list[0][0][1]
# [] is used to call out the specific element. 
# in python, the first element is '0', unlike in R.
# calling out test_list[0] will give the only element
# which is [[1,2,3]]. repeat to call out [1,2,3]
```
