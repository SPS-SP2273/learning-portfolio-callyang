## 1.3 Accessing data from a dictionary

```python
superhero_info = {
    "Natasha Romanoff": "Black Widow",
    "Tony Stark": "Iron Man",
    "Stephen Strange": "Doctor Strange"
}             

superhero_info["Natasha Romanoff"]
```

```python
superhero_info["Doctor Strange"]

# will not work, cannot ask python what is the key for this value
# only can ask python what is the value of this key
# dictionary is key:value structure
```

```python
print(superhero_info.keys())

# .keys() to find keys

print(superhero_info.values())

# .values() to find values
```

## 1.4 Higher dimensional lists

```python
# In dictionary, we can store key:value easily
# However for lists, we need to separate each key:value 
# as its own list with []

py_superhero = [['Natasha Romanoff', 'Black Widow'],
               ['Tony Stark','Iron Man'],
               ['Stephen Strange','Doctor Strange']]
print(py_superhero)
print(py_superhero[2])    # gives the THIRD!! element,
                          # which itself is a list 
print(py_superhero[2][0])

listt = [[[["hi"]]]]
listt[0][0][0][0]    # remember python first element index is 0
```

# 2 Lists vs. Arrays

## 2.1 Size

```python
numberlist = [[1,"A"],
             [2,"B"],
             [3,"C"],
             [4,"D"]
             ]

numberarray = np.array(numberlist)
print(numberarray)
print(numberlist)

# after printing out, notice difference in format

print(len(numberlist)) # len() tells u there are 4 elements in this list
print(len(numberarray))
print(numberarray.shape)

# .shape tells you 'shape' of array, 4 row by 2 column
# no such thing for list as list is just one row

# the .shape or .keys .values is not function, 
# is simply calling out python
# to give u the attirbute of the array/dict/list
```

## 2.2 Arrays are fussy about type

```python
# data types
# int, float, str
# Array can only have ONE data type
```

```python
listexample = [1,2, 'hehe']
print(listexample)

arrayexample = np.array(listexample)
print(arrayexample)

# as long as there is ONE string, everyth gets converted 
# to string in python array
```

## 2.3 Adding a number

```python
simplelist = [1,2,3]
simplearray = np.array(simplelist) # using list to create array
                                   # can do manually
                                   # but remember ([])

simplelist + 10 # will not work
simplearray + 10 # notice that since the prev line doesnt work
                 # python will not run the next line
```

```python
simplearray + 10     # +10 to each element
```

## 2.4 Adding another list

```python
# however, u can add a list to another list
# to make one longgg list

simplelist = [1,2,3,4,5]
ssecondlist = [2,3,4,5,]

simplelist + ssecondlist 
```

```python
# u can also add arrays

simplearray = np.array(simplelist)
ssecondarray = np.array(ssecondlist)

simplearray + ssecondarray
# will not work because the length is not equal
```

```python
thirdarray = np.array([3,4,5,5,5])
simplearray + thirdarray
```
